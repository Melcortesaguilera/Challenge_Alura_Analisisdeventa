# 5. Envío promedio por tienda

```python
#envio promedio tienda 1
envio_t1 = tienda['Costo de envío'].mean()
envio_t1 = round(envio_t1, 2)
print(f'La Tienda 1 presenta un costo de envío promedio de ${envio_t1}')
```

```python
#envio promedio tienda 2
envio_t2 = tienda2['Costo de envío'].mean()
envio_t2 = round(envio_t2, 2)
print(f'En promedio, los envíos de la Tienda 2 cuestan ${envio_t2}')
```

```python
#envio promedio tienda 3
envio_t3 = tienda3['Costo de envío'].mean()
envio_t3 = round(envio_t3, 2)
print(f'Cada envío en la Tienda 3 tiene un costo promedio de ${envio_t3}')
```

```python
#envio promedio tienda 4
envio_t4 = tienda4['Costo de envío'].mean()
envio_t4 = round(envio_t4, 2)
print(f'Cada envío en la Tienda 4 tiene un costo promedio de ${envio_t4}')
```
