#1. Análisis de facturación



```python
#ingreso total de cada tienda.
total_ingreso_tienda1 = tienda['Precio'].sum()
print(f'El ingreso acumulado para la Tienda 1 es de ${total_ingreso_tienda1}')
```

```python
total_ingreso_tienda2 = tienda2['Precio'].sum()
print(f'La Tienda 2 generó un ingreso total de ${total_ingreso_tienda2}')
```

# 2. Ventas por categoría

```python
#ventas por categoria tienda 1
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria_tienda1 = ventas_por_categoria_tienda1.sort_values(ascending=False) #para ver la lista en orden desendente
print(ventas_por_categoria_tienda1)
```

```python
#ventas por categoria tienda 2
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria_tienda2 = ventas_por_categoria_tienda2.sort_values(ascending=False) #para ver la lista en orden desendente
print(ventas_por_categoria_tienda2)
```

```python
#ventas por categoria tienda 3
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria_tienda3 =ventas_por_categoria_tienda3.sort_values(ascending=False) #para ver la lista en orden desendente
print(ventas_por_categoria_tienda3)
```

```python
#ventas por categoria tienda 4
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria_tienda4 = ventas_por_categoria_tienda4.sort_values(ascending=False) #para ver la lista en orden desendente
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
#calificacion promedio por tienda 1
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda1 = round(calificacion_tienda1, 2)
print(f'La calificación promedio de los clientes para la Tienda 1 es de {calificacion_tienda1}')
```

```python
#calificacion promedio tienda 2
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda2 = round(calificacion_tienda2, 2)
print(f'La calificación promedio de los clientes para la Tienda 2 es de {calificacion_tienda2}')
```

```python
#calificacion promedio tienda 3
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda3 = round(calificacion_tienda3, 2)
print(f'En la Tienda 3, los productos recibieron una calificación promedio de {calificacion_tienda3}')
```

```python
#calificacion promedio tienda 4
calificacion_tienda4 = tienda4['Calificación'].mean()
calificacion_tienda4 = round(calificacion_tienda4,)
print(f'La Tienda 4 obtuvo una puntuación promedio de satisfacción de {calificacion_tienda4}')
```

# 5. Envío promedio por tienda
